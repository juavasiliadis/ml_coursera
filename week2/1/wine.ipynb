```python
from sklearn.datasets import load_boston

from sklearn.neighbors import KNeighborsRegressor
bos = load_boston()
X = bos['data']
y = bos['target']


import sklearn.preprocessing as pr
import numpy as np

X = pr.scale(X, axis=0)
p = np.linspace(1, 10, num=200)

result={}

for i in p:
    neigh = KNeighborsRegressor(n_neighbors=5, weights='distance',p=i)
    neigh.fit(X,y)
    kf = KFold(n_splits=5,random_state = 42, shuffle = True)
    kf.get_n_splits(X)
    cv_results = cross_val_score(neigh,X,y,scoring='neg_mean_squared_error')
    result[i] = cross_val_score(neigh, X, y, cv=kf)
    

for k in result.keys():
    result[k] = np.mean(result[k])

for k in result.keys():
    print(str(k) + ' ' + str(result[k]))

print(max(result.values()))
print(list(result.keys())[list(result.values()).index(max(result.values()))])

```
